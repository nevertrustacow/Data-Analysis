## Q1.3
Getting the median for the columns and
replacing all the missing values with the median

```python
#calculating median for "Rtn..1.Day",	"Rtn..1.Day.1"	and Rtn..2.Day columns
median_Rtn_1_Day = df['Rtn..1.Day'].median()
median_Rtn_1_Day_1 = df['Rtn..1.Day.1'].median()
median_Rtn_2_Day = df['Rtn..2.Day'].median()
```

```python
#replacing missing values on these columns with median values
df['Rtn..1.Day'][(df['Rtn..1.Day']).isna()] = median_Rtn_1_Day
df['Rtn..1.Day.1'][(df['Rtn..1.Day.1']).isna()] = median_Rtn_1_Day_1
df['Rtn..2.Day'][(df['Rtn..2.Day']).isna()] = median_Rtn_2_Day
```

```python
#Displaying the new dataset
df.head(10)
```

```python
#Display the new summary of the dataset 
df.info()
```

## Q1.4
Create a new dataframe named “summary” 

```python
#Create a new dataframe named “summary” and group the following columns together: Date, Rtn..1.Day, Rtn..1.Day.1
#Rtn..2.Day, Count the column Count 
cols = ['Date', 'Rtn..1.Day', 'Rtn..1.Day.1', 'Rtn..2.Day']
summary = pd.DataFrame(df.groupby(cols)['Close'].count())
summary.columns = ['Count']
#Show a summary of the newly created dataframe
summary.info()
```

```python
#display a summary of the new table
summary.info()
```

## Q1.6
Convert the “Date” column from the “df” dataset and “summary” dataframe

```python
#Convert the “Date” column from the “df” dataset and “summary” dataframe
df['Date'] = pd.to_datetime(df['Date'].str.replace('/','-'))
summary['Date'] = pd.to_datetime(summary['Date'].str.replace('/','-'))
```

```python
#Extract the year, month and day from the Date columnn
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

summary['Year'] = summary['Date'].dt.year
summary['Month'] = summary['Date'].dt.month
summary['Day'] = summary['Date'].dt.day
```

```python
# Display the updated dataset in tabular format. 
summary.head(10)
```

## Q1.7
Create a countplot that will display the total shares invested for each year

```python
plt.figure(figsize = (10, 10))
#putting the years in the variable x
x = summary.Year.value_counts().index
#putting the number of shares in the variable y
y = summary.Year.value_counts().values
#creating the count plot
plt.bar(x, y)
plt.xlabel('Years') #setting x label to 'Years'
plt.ylabel('Number of Shares') #setting y label to 'Number of Shares'
plt.title('Total Shares Per Year') #setting the title to 'Total Shares Per Year'
plt.xticks(x)
plt.ylim(240, 255) #Setting the y-axis starting value at 240 and the ending value at 255. 
plt.show()
```

- the number of invested shares bewteen 2000 and 2017 is flactuating picking twice in 2004 and 2010 with total shares of 254 and 252 respectively.
- 2000 and 2017 had the lowest number of shares not reaching 240 shares per year.


## Q1.8
Creating a lineplot that will indicate each share’s value from each record by using
the following columns’ data: Rtn..2.Day, Rtn..1.Day.1 

```python
x = summary.index # the record number
line1 = summary['Rtn..1.Day.1']
line2 = summary['Rtn..2.Day']
plt.figure(figsize = (15, 10))
plt.plot(x, line1, c = 'r', label = 'Rtn..1.Day.1', alpha = .5) #lineplot for the 'Rtn..1.Day.1' column
plt.plot(x, line2, c = 'b', label = 'Rtn..2.Day', alpha = .5) #lineplot for the 'Rtn..2.Day' column
plt.xlabel('Record Number') #setting x label to 'Record Number'
plt.ylabel('Share Value') #setting x label to 'Share Value'
plt.title('Shares Value Per Record ') #setting the title to 'Shares Value Per Record '
plt.legend()
plt.show()
```

- we can see that the Rtn.2Day has a wider range of variation comparing to Rtn.1.Day.1
