```python
import csv
import random 

#fuc to gen random data 
    
def gen_dataset(filename,rows):
    with open(filename + ".csv", "w", newline="") as csvfile:
        fieldnames = ["PassengerID", "Pclass","Age", "Gender", "Survived"]
        csv_writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        csv_writer.writeheader()
        
        for _ in range (rows):
            PassengerID = random.randint(1,1000)
            Age = random.randint(1, 80)
            Pclass = random.choice([1,2,3])
            Gender = random.choice(["Male", "Female"])
            Survived = random.choice(["Yes", "No"])
            
            
            csv_writer.writerow({
                "PassengerID": PassengerID,
                "Pclass": Pclass,
                "Age": Age,
                "Gender": Gender,
                "Survived": Survived              
                
            })
            
    
    
gen_dataset("Titanic_data", 1000)

print("Titanic_data, successfully created!")
```

```python
#cleaning the data 
import pandas as pd 

#loading it 
Titanic_data = pd.read_csv("Titanic_data.csv")

print("Initial dataset statistics: ")
print("Titanic_data: ")
print(Titanic_data.describe())

# cleaning it/ removing null values 
Titanic_data_cleaned = Titanic_data[(Titanic_data != 0).all(1)]

print("Cleaned dataset statistics: ")
print("Titanic_data_cleaned: ")
print(Titanic_data_cleaned.describe())

Titanic_data_cleaned.to_csv("Titanic_data_cleaned.csv", index=False)

print("\nDataset has now been cleaned andd saved to 'Titanic_data_cleaned'")
```

```python
# creating the frequency table 
import pandas as pd

Titanic_data_cleaned = pd.read_csv("Titanic_data_cleaned.csv")

#freq for those who survived 
Survive_freq = Titanic_data_cleaned["Survived"].value_counts()

#freq age 
age_freq = Titanic_data_cleaned["Age"].value_counts()

# freq class 

pClass_freq = Titanic_data_cleaned["Pclass"].value_counts()

gender_freq = Titanic_data_cleaned["Gender"].value_counts()

print("Frequency table for those who survived: ")
print(Survive_freq)

print("Frequency table for age: ")
print(age_freq)

print("Frequency of passenger class: ")
print(pClass_freq)

print("Frequency of Genders: ")
print(gender_freq)
```
